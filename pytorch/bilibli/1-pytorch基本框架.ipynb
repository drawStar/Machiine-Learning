numpy实现两层神经网络

一个隐藏层，没有bias

$h=W_1X$

$a=ReLU(h)=max(0,h)$

$\hat y = W_2a$

$MSELoss= \frac{1}{N}\sum (y - \hat y)^2$


```
import numpy as np

#y样本数，输入的dim，hidden层神经元个数，输出Dim
N, D_in, H, D_out = 64,1000,100,10 

#随机创建一些训练数据
x = np.random.randn(N,D_in)
y = np.random.randn(N,D_out) 
w1 = np.random.randn(D_in, H)
w2 = np.random.randn(H, D_out)

learningRate = 1e-6

for epochIndex in range(500):
    #forward pass
    h = x.dot(w1) # [N,H]
    a = np.maximum(h, 0) 
    y_pred = a.dot(w2) #[N, D_out]

    #compuate MSE loss
    loss = (np.square(y-y_pred).sum()) / N
    print(epochIndex,loss)

    #backward pass
    #计算梯度,我先不管转不转置了
    grad_y_pred =2.0/N*(y_pred-y) #Loss对ypred求导
    grad_w2 = a.T.dot(grad_y_pred) #Loss对w2求导
    grad_a = grad_y_pred.dot(w2.T)
    grad_h = a.copy()#<0的导数是0，>0部分还是本身（相当于乘以1.所以copy）
    grad_h[h<0] = 0
    grad_w1 = x.T.dot(grad_h)

    #update weight
    w1 = w1 - learningRate * grad_w1
    w2 = w2 - learningRate * grad_w2

# 可以看出loss在下降
```

使用Pytorch的Tensor来实现

np.random.randn改成torch.randn

dot改为mm（矩阵相乘）

torch.relu

loss要加item（）


```
import numpy as np

#y样本数，输入的dim，hidden层神经元个数，输出Dim
N, D_in, H, D_out = 64,1000,100,10 

#随机创建一些训练数据
x = torch.randn(N,D_in)
y = torch.randn(N,D_out) 
w1 = torch.randn(D_in, H)
w2 = torch.randn(H, D_out)

learningRate = 1e-6

for epochIndex in range(500):
    #forward pass
    h = x.mm(w1) # [N,H]
    a = torch.relu(h) 
    y_pred = a.mm(w2) #[N, D_out]

    #compuate MSE loss
    loss = (((y-y_pred).pow(2).sum()) / N)
    print(epochIndex,loss.item())

    #backward pass
    #计算梯度,我先不管转不转置了
    grad_y_pred =2.0/N*(y_pred-y) #Loss对ypred求导
    grad_w2 = a.t().mm(grad_y_pred) #Loss对w2求导
    grad_a = grad_y_pred.mm(w2.t())
    grad_h = a.clone()#<0的导数是0，>0部分还是本身（相当于乘以1.所以copy）
    grad_h[h<0] = 0
    grad_w1 = x.t().mm(grad_h)

    #update weight
    w1 = w1 - learningRate * grad_w1
    w2 = w2 - learningRate * grad_w2

# 可以看出loss在下降
```

使用Pytorch自动求导

在要求导的变量后写requires_grad=True

loss.backward()


```
import numpy as np

#y样本数，输入的dim，hidden层神经元个数，输出Dim
N, D_in, H, D_out = 64,1000,100,10 

#随机创建一些训练数据
x = torch.randn(N,D_in)
y = torch.randn(N,D_out) 
w1 = torch.randn(D_in, H, requires_grad=True)
w2 = torch.randn(H, D_out, requires_grad=True)

learningRate = 1e-6

for epochIndex in range(500):
    #forward pass
    h = x.mm(w1) # [N,H]
    a = torch.relu(h) 
    y_pred = a.mm(w2) #[N, D_out]

    #compuate MSE loss
    loss = (((y-y_pred).pow(2).sum()) / N)
    print(epochIndex,loss.item())

    #backward pass


    loss.backward() 

    #update weight
    with torch.no_grad(): #不追踪导数，不用于W1,W2导数的计算
        w1.sub_(w1.grad*learningRate)
        #注意要使用inplace操作，不能w1-=w1.grad*lr,这样w1会指向一个全新的tensor，原始tensor信息会丢失
        #https://discuss.pytorch.org/t/nonetype-object-has-no-attribute-zero/61013/2
        w2.sub_(w2.grad*learningRate)
        w1.grad.zero_() #每个epoch清零梯度缓存，否则会累积
        w2.grad.zero_() #清零梯度缓存


# 可以看出loss在下降
```

```
import numpy as np

#y样本数，输入的dim，hidden层神经元个数，输出Dim
N, D_in, H, D_out = 64,1000,100,10 

#随机创建一些训练数据
x = torch.randn(N,D_in)
y = torch.randn(N,D_out) 
model = torch.nn.Sequential(
    torch.nn.Linear(D_in, H),#注意这里加了bias，w1*x+b
    torch.nn.ReLU(),
    torch.nn.Linear(H, D_out)
)

#初始化权重，可选
torch.nn.init.normal_(model[0].weight)#正态分布初始化第一个线性层的权重
torch.nn.init.normal_(model[2].weight)#正态分布初始化第二个线性层的权重

learningRate = 1e-6
loss_fn = torch.nn.MSELoss(reduction='mean')

for epochIndex in range(500):
    #forward pass
    # h = x.mm(w1) # [N,H]
    # a = torch.relu(h) 
    # y_pred = a.mm(w2) #[N, D_out]
    y_pred = model(x)


    #compuate MSE loss
    loss = loss_fn(y_pred, y)
    print(epochIndex,loss.item())

    #backward pass

    model.zero_grad()#下一次求导前，清零缓存,参数封装到model里了

    loss.backward() 

    #update weight
    with torch.no_grad(): #不追踪导数，不用于W1,W2导数的计算
        for parm in model.parameters():
            parm.sub_(parm.grad*learningRate)


# 可以看出loss在下降
```

pytorch optim

不再手动更新模型的weights，而是使用optim这个包来帮助我们更新参数，

```
import numpy as np

#y样本数，输入的dim，hidden层神经元个数，输出Dim
N, D_in, H, D_out = 64,1000,100,10 

#随机创建一些训练数据
x = torch.randn(N,D_in)
y = torch.randn(N,D_out) 
model = torch.nn.Sequential(
    torch.nn.Linear(D_in, H),#注意这里加了bias，w1*x+b
    torch.nn.ReLU(),
    torch.nn.Linear(H, D_out)
)

#初始化权重，可选
torch.nn.init.normal_(model[0].weight)#正态分布初始化第一个线性层的权重
torch.nn.init.normal_(model[2].weight)#正态分布初始化第二个线性层的权重

learningRate = 1e-4
loss_fn = torch.nn.MSELoss(reduction='mean')
optimizer = torch.optim.Adam(model.parameters(), lr=learningRate)

for epochIndex in range(500):
    #forward pass
    # h = x.mm(w1) # [N,H]
    # a = torch.relu(h) 
    # y_pred = a.mm(w2) #[N, D_out]
    y_pred = model(x)


    #compuate MSE loss
    loss = loss_fn(y_pred, y)
    print(epochIndex,loss.item())

    #backward pass
    optimizer.zero_grad()

    loss.backward() 

    # update model parameters
    optimizer.step()

# 可以看出loss在下降
```

Pytorch:自定义nn.Modules


```
import numpy as np

#y样本数，输入的dim，hidden层神经元个数，输出Dim
N, D_in, H, D_out = 64,1000,100,10 

#随机创建一些训练数据
x = torch.randn(N,D_in)
y = torch.randn(N,D_out) 

class TwoLayerNet(torch.nn.Module):
    def __init__(self, D_in, H, D_out):
        super(TwoLayerNet, self).__init__()
        self.Linear1 = torch.nn.Linear(D_in, H)
        self.Linear2 = torch.nn.Linear(H, D_out)

    def forward(self, x):
        y_pred = self.Linear2(torch.relu(self.Linear1(x)))
        return y_pred #别忘了return

model = TwoLayerNet(D_in, H, D_out)
learningRate = 1e-4
loss_fn = torch.nn.MSELoss(reduction='mean')
optimizer = torch.optim.Adam(model.parameters(), lr=learningRate)

for epochIndex in range(500):
    y_pred = model(x)
    #compuate MSE loss
    loss = loss_fn(y_pred, y)
    print(epochIndex,loss.item())

    #backward pass
    optimizer.zero_grad()

    loss.backward() 

    # update model parameters
    optimizer.step()

# 可以看出loss在下降
```
